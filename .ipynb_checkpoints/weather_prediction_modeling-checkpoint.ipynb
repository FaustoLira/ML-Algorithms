## Spliting the data

Splitting the data into training and testing sets with a ratio of 80% and 20%, respectively.

```python
# split the data into features and target
temp_seol_features = temp_seol.drop(columns=['NextDayAvTemp'])
temp_seol_target = temp_seol['NextDayAvTemp']

# Split the dataset into train_temp and test sets
train_features, test_features, train_target, test_target =\
    train_test_split(temp_seol_features, temp_seol_target,\
                     test_size=0.2, random_state=123)

# reset index
train_features.reset_index(drop=True, inplace=True)
test_features.reset_index(drop=True, inplace=True)
train_target.reset_index(drop=True, inplace=True)
test_target.reset_index(drop=True, inplace=True)
```

## One-Hot-Encoding
   
1. Convert 'station' to categorical variable
2. Transform using Dummy Variable (drop_first=True)

```python
# Transform station into dummy variable

# convert to string
train_features_imputed['station'] = train_features_imputed['station'].astype('str').str.replace('.0', '')
test_features_imputed['station'] = test_features_imputed['station'].astype('str').str.replace('.0', '')

# one-hot-encoding
enc = OneHotEncoder(handle_unknown='ignore', sparse_output=False, drop='first')

# training
train_features_imputed_cat = enc.fit_transform(train_features_imputed[['station']])
train_features_cat = pd.DataFrame(train_features_imputed_cat, columns=enc.get_feature_names_out(['station']))

# sorting columns
cols_sorted = [f'station_{i}' for i in range(2, 26)]
train_features_cat = train_features_cat[cols_sorted]

# testing
test_features_imputed_cat = enc.transform(test_features_imputed[['station']])
test_features_cat = pd.DataFrame(test_features_imputed_cat, columns=enc.get_feature_names_out(['station']))

# sorting columns
cols_sorted = [f'station_{i}' for i in range(2, 26)]
test_features_cat = test_features_cat[cols_sorted]
```

```python
# concat the data
train_features_transformed = pd.concat([train_features_cat, train_features_imputed.drop(columns=['station'])], axis=1)
test_features_transformed = pd.concat([test_features_cat, test_features_imputed.drop(columns=['station'])], axis=1)
```

## Machine Learning Pipeline

```python
results_df = pd.DataFrame({
    'scaler': [],
    'feat_sel name': [],
    '# features': [],
    'model': [],
    'r_square_train': [],
    'r_square_test': []
})
```

```python
# All scalers to be used used
scalers = {
    'MinMaxScaler' : MinMaxScaler(), 
    'Normalizer': Normalizer(), 
    'RobustScaler': RobustScaler(), 
    'StandardScaler': StandardScaler()
}

# All possible number of features to be used
k_values = [15, 30, 45]

# All score functions to be used
score_functions = {
    'f_regression': f_regression, 
    'r_regression': r_regression
}

# All ML models to be used
models_dict = {
    'XGBRegressor': XGBRegressor(n_jobs=-1),
    'LinearRegression': LinearRegression(n_jobs=-1),
    'Ridge': Ridge(),
    'Lasso': Lasso(),
    'ElasticNet': ElasticNet(),
    'DecisionTreeRegressor': DecisionTreeRegressor(),
    'RandomForestRegressor': RandomForestRegressor(n_jobs=-1),
    'SVR': SVR(),
    'MLPRegressor': MLPRegressor(),
    'KNeighborsRegressor': KNeighborsRegressor(n_jobs=-1),
    'AdaBoostRegressor': AdaBoostRegressor(),
    'GaussianProcessRegressor': GaussianProcessRegressor()
}
```

```python
import warnings
from sklearn.exceptions import ConvergenceWarning
warnings.filterwarnings('ignore', category=DeprecationWarning)
warnings.filterwarnings('ignore', category=ConvergenceWarning)

index = 0

for sc_name, scaler in scalers.items():
    for k in k_values:
        for sf_name, score_function in score_functions.items():
            for m_name, model in models_dict.items():
                # Instantiate pipeline
                pipe = Pipeline([
                    (sc_name, scaler),
                    (sf_name, SelectKBest(score_function, k=k)),
                    (m_name, model)
                ])

                # Fit pipeline
                pipe.fit(train_features_transformed, train_target)
                
                # Predict on training set
                y_pred_train = pipe.predict(train_features_transformed)
                
                # Calculate R^2 score for training set
                r2_train = r2_score(train_target, y_pred_train)

                # Predict on test set
                y_pred_test = pipe.predict(test_features_transformed)
                
                # Calculate R^2 score for test set
                r2_test = r2_score(test_target, y_pred_test)
                
                results_df.loc[index] = [sc_name, sf_name, k, m_name, r2_train, r2_test]

                index = index + 1
```
